### In this notebook, we analyze chocolate sales data to uncover trends and forecast future sales.  
Our objective is to predict chocolate demand over time, helping optimize production and logistics, forecasting sales for the next 30 days.

### Models Used
- Prophet
- ARIMA
- SARIMA
- XGBoost

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session

import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
import warnings
warnings.filterwarnings("ignore")
```

## Data Overview

```python
df=pd.read_csv("/kaggle/input/chocolate-sales/Chocolate Sales.csv")
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y')
df['Amount'] = df['Amount'].replace('[\$,]', '', regex=True).astype(float)
df.head()
```

## Aggregate Sales by Date

```python
daily_sales = df.groupby('Date')['Amount'].sum().reset_index()
daily_sales.sort_values('Date', inplace=True)
daily_sales.head()
```

## Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Amount', data=daily_sales)
plt.title('Daily Chocolate Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount ($)')
plt.show()
```

## Forecasting with Prophet

```python
prophet_df = daily_sales.rename(columns={'Date': 'ds', 'Amount': 'y'})
model = Prophet()
model.fit(prophet_df)
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)
forecast_prophet = forecast.tail(30)[['ds', 'yhat']].rename(columns={'ds': 'Date', 'yhat': 'Prophet'})
fig = model.plot(forecast)
plt.title('Chocolate Sales Forecast (Next 30 Days)')
plt.show()
```

## Forecasting with ARIMA

```python
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from sklearn. metrics import mean_squared_error

df_arima = daily_sales.set_index('Date').asfreq('D').fillna(0)['Amount']

model_arima = ARIMA(df_arima, order=(1,1,1))
result_arima = model_arima.fit()

forecast_arima = result_arima.get_forecast(steps=30)
forecast_mean = forecast_arima.predicted_mean
forecast_ci = forecast_arima.conf_int()

forecast_dates = pd.date_range(start=df_arima.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')

plt.figure(figsize=(12,6))
plt.plot(df_arima, label='Observed')
plt.plot(forecast_dates, forecast_mean.values, label='ARIMA Forecast', color='orange')
plt.fill_between(forecast_dates, forecast_ci.iloc[:,0], forecast_ci.iloc[:,1], color='orange', alpha=0.2)
plt.title('Chocolate Sales Forecast with ARIMA (Next 30 Days)')
plt.xlabel('Date')
plt.ylabel('Sales Amount ($)')
plt.legend()
plt.show()
```

## Forecasting With SARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
sarima_model = SARIMAX(df_arima, order=(1,1,1), seasonal_order=(1,1,1,7))  # Weekly seasonality example
sarima_result = sarima_model.fit()
sarima_forecast = sarima_result.get_forecast(steps=30)
sarima_mean = sarima_forecast.predicted_mean
sarima_ci = sarima_forecast.conf_int()
plt.figure(figsize=(12,6))
plt.plot(df_arima, label='Observed')
plt.plot(forecast_dates, sarima_mean.values, label='SARIMA Forecast', color='green')
plt.fill_between(forecast_dates, sarima_ci.iloc[:,0], sarima_ci.iloc[:,1], color='green', alpha=0.2)
plt.title('Chocolate Sales Forecast with SARIMA (Next 30 Days)')
plt.xlabel('Date')
plt.ylabel('Sales Amount ($)')
plt.legend()
plt.show()
```

## Forecasting With XGBRegressor

```python
from xgboost import XGBRegressor
lag = 7  # use previous 7 days as features
xgb_data = df_arima.reset_index()
for i in range(1, lag+1):
    xgb_data[f'lag_{i}'] = xgb_data['Amount'].shift(i)
xgb_data.dropna(inplace=True)
X = xgb_data[[f'lag_{i}' for i in range(1, lag+1)]]
y = xgb_data['Amount']
model_xgb = XGBRegressor(objective='reg:squarederror', n_estimators=100)
model_xgb.fit(X, y)
recent = df_arima[-lag:].values.reshape(1, -1)
xgb_preds = []
for _ in range(30):
    pred = model_xgb.predict(recent)[0]
    xgb_preds.append(pred)
    recent = np.roll(recent, -1)
    recent[0, -1] = pred
xgb_forecast_dates = pd.date_range(start=df_arima.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')
plt.figure(figsize=(12,6))
plt.plot(df_arima, label='Observed')
plt.plot(xgb_forecast_dates, xgb_preds, label='XGBoost Forecast', color='red')
plt.title('Chocolate Sales Forecast with XGBoost (Next 30 Days)')
plt.xlabel('Date')
plt.ylabel('Sales Amount ($)')
plt.legend()
plt.show()
```

## Summary 

```python
summary_df = pd.DataFrame({
    'Date': forecast_prophet['Date'].values,
    'Prophet': forecast_prophet['Prophet'].values,
    'ARIMA': forecast_mean.values,
    'SARIMA': sarima_mean.values,
    'XGBoost': xgb_preds
})
summary_df = summary_df.round(2)
display(summary_df.head(20))
```

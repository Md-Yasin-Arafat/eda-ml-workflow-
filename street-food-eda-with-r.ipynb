```r
library(readr) 
library(dplyr) 
library(ggplot2)

df <- read_csv("/kaggle/input/global-street-food-dataset/global_street_food.csv")
head(df)
```

```r
glimpse(df)
str(df)
```

## **Distribution of coocking method**

```r
options(repr.plot.width = 16, repr.plot.height = 10) 

ggplot(df, aes(x = `Cooking Method`)) +
  geom_bar(fill = "steelblue", color = "black") +
  labs(title = "Distribution of Cooking Methods", x = "Cooking Method", y = "Count") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(size = 16, angle = 45, hjust = 1),  
    axis.text.y = element_text(size = 16), 
    axis.title.x = element_text(size = 18), 
    axis.title.y = element_text(size = 18), 
    plot.title = element_text(size = 20, face = "bold")  
  )
```

## **Distribution of Price**

```r
ggplot(df, aes(x = `Typical Price (USD)`)) +
  geom_histogram(binwidth = 0.2, fill = "steelblue", color = "black") +
  labs(title = "Distribution of Street Food Prices", x = "Price (USD)", y = "Count") +
  theme_minimal()+
  theme(
    axis.text.x = element_text(size = 16, angle = 45, hjust = 1),  
    axis.text.y = element_text(size = 16), 
    axis.title.x = element_text(size = 18), 
    axis.title.y = element_text(size = 18), 
    plot.title = element_text(size = 20, face = "bold")  
  )
```

## **Vegetarian vs Non-Vegetarian Street Foods**

```r
ggplot(df, aes(x = Vegetarian)) +
  geom_bar(fill = "skyblue", color = "black") +
  labs(title = "Vegetarian vs Non-Vegetarian Street Foods", x = "Vegetarian", y = "Count") +
  theme_minimal()+
  theme(
    axis.text.x = element_text(size = 16, angle = 45, hjust = 1),  
    axis.text.y = element_text(size = 16), 
    axis.title.x = element_text(size = 18), 
    axis.title.y = element_text(size = 18), 
    plot.title = element_text(size = 20, face = "bold")  
  )
```

## **Price Distribution by Country**

```r
ggplot(df, aes(x = Country, y = `Typical Price (USD)`, fill = Country)) +
  geom_boxplot() +
  labs(title = "Price Distribution by Country", x = "Country", y = "Price (USD)") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))+
  theme(
    axis.text.x = element_text(size = 16, angle = 45, hjust = 1),  
    axis.text.y = element_text(size = 16), 
    axis.title.x = element_text(size = 18), 
    axis.title.y = element_text(size = 18), 
    plot.title = element_text(size = 20, face = "bold")  
  )
```

## **Most Common Ingredients**

```r
library(tidytext)

df_ingredients <- df %>%
  unnest_tokens(word, Ingredients, token = "words") %>%
  count(word, sort = TRUE)

ggplot(df_ingredients %>% top_n(15), aes(x = reorder(word, n), y = n)) +
  geom_col(fill = "tomato") +
  coord_flip() +
  labs(title = "Most Common Ingredients in Global Street Food", x = "Ingredient", y = "Count") +
  theme_minimal()+
  theme(
    axis.text.x = element_text(size = 16, angle = 45, hjust = 1),  
    axis.text.y = element_text(size = 16), 
    axis.title.x = element_text(size = 18), 
    axis.title.y = element_text(size = 18), 
    plot.title = element_text(size = 20, face = "bold")  
  )
```

## **Common Foods' Distribution**

```r
df %>%
  count(`Dish Name`, sort = TRUE) %>%
  top_n(10) %>%
  ggplot(aes(x = reorder(`Dish Name`, n), y = n)) +
  geom_col(fill = "dodgerblue") +
  coord_flip() +
  labs(title = "Top 10 Most Common Street Foods", x = "Dish Name", y = "Count") +
  theme_minimal()+
  theme(
    axis.text.x = element_text(size = 16, angle = 45, hjust = 1),  
    axis.text.y = element_text(size = 16), 
    axis.title.x = element_text(size = 18), 
    axis.title.y = element_text(size = 18), 
    plot.title = element_text(size = 20, face = "bold")  
  )
```

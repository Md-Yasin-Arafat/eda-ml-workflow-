```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")  # Suppress all warnings
```

```python
df=pd.read_csv("/kaggle/input/car-dataset/all_cars_datset_final.csv")
```

## Get an Overview of the Dataset

```python
df.head(25)
```

```python
df.rename(columns={'FY2024(sales)': 'Sales'}, inplace=True)
```

```python
print(df.info())
print(df.describe(include='all'))
```

## Clean Columns

```python
df['Price'] = df['Price'].str.replace('Rs\.|Lakh|Crore|,', '', regex=True).str.strip()
df['Price'] = df['Price'].apply(
    lambda x: float(x) * 100 if 'Crore' in x else float(x) if x.replace('.', '').isdigit() else np.nan
)

df['Rating'] = df['Rating'].str.replace('04-May|05-May|03-May', '', regex=True)
df['Rating'] = df['Rating'].str.extract(r'(\d+\.?\d*)')[0].astype(float)

df['Safety'] = df['Safety'].str.extract(r'(\d+)').astype(float)

df['Mileage'] = df['Mileage'].str.extract(r'(\d+\.?\d*|\d+)').astype(float)

df['Sales'] = df['Sales'].str.replace(',', '').replace('', np.nan).astype(float)
```

## Impute missing numerical values

```python
df = df.dropna(subset=['Price', 'Rating','Sales','Power (BHP)'])

df['Safety'] = df['Safety'].fillna(df['Safety'].median())
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].median())
```

## Deal with Messy Column

```python
def extract_avg_bhp(value):
    value = value.replace('bhp', '').strip()
    if '-' in value:
        low, high = map(float, value.split('-'))
        return (low + high) / 2
    else:
        return float(value)

df['Power (BHP)'] = df['Power (BHP)'].apply(extract_avg_bhp)
```

```python
df.rename(columns={'Power (BHP)': 'Average Power (BHP)'}, inplace=True)
```

## Top Revenue-Generating Brands

```python
df['Total_Revenue'] = df['Sales'] * df['Price']
```

```python
# Group by Brand and sum the Total_Revenue
brand_revenue = df.groupby("Brand", as_index=False)["Total_Revenue"].sum()

# Sort and get top 10 brands
top_brands = brand_revenue.sort_values(by="Total_Revenue", ascending=False).head(10)

# Plotting
plt.figure(figsize=(10, 8))
sns.barplot(data=top_brands, x="Brand", y="Total_Revenue", palette="Set2")
plt.title("Top 10 Brands by Total Revenue")
plt.ylabel("Total Revenue")
plt.xlabel("Brands")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Price vs. Sales

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Price', y='Sales', data=df, hue='Mileage',palette="viridis")
plt.title('Price vs. Sales')
plt.show()
```

## Correlation heatmap

```python

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='PiYG')
plt.title('Correlation Matrix')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

## Conclusion from the Heatmap

- Sales decrease as **Power** increases, with a correlation of approximately **-45%**.  
- Sales also decline with higher **Price** (**-33%**) and **Safety** ratings (**-46%**).  
- Conversely, sales tend to increase with better **Mileage** and higher **Rating** scores.
